### Zadnie 5: Analiza regresji [25 pkt]

```python
# Tabela
# YOUR CODE HERE

col1 = pd.MultiIndex.from_product([daily_bike_traffic_data.columns, ['a','b']])
col2 = daily_bike_traffic_data.columns
indx = weather_data.columns
regression_data_1 = pd.DataFrame(columns=col1, index=indx)

for i in indx:
    for c in col2:
        df = pd.DataFrame({'x':weather_data[i], 'y':daily_bike_traffic_data[c]}).dropna()
        m = sp.stats.linregress(df['x'], df['y'])
        regression_data_1[c,'a'][i] = m.slope
        regression_data_1[c,'b'][i] = m.intercept


regression_data_1.style.format(precision=3)
```

```python
### Komórka testowa

# Testy wybranych wartości z tablicy regression_data_1
assert np.isclose(regression_data_1.loc[("Temperature", "Avg"), ('Total', 'a')], 118.37455627981312)
assert np.isclose(regression_data_1.loc[("Snow", "Depth"), ('Total', 'b')], 2219.810729613734)
assert np.isclose(regression_data_1.loc[("Wind", "Avg speed"), ('East', 'a')], -74.72308463717329)
assert np.isclose(regression_data_1.loc[("Precipitation", "Total"), ('East', 'b')], 956.7960528263079)
assert np.isclose(regression_data_1.loc[("Temperature", "Min"), ('West', 'a')], 77.54619929276554)
assert np.isclose(regression_data_1.loc[("Wind", "Max speed"), ('West', 'b')], 1970.6955983993344)
```

```python
# Wykresy
# YOUR CODE HERE
fig, ax = plt.subplots(3,3,figsize=(18,18))
fig.suptitle('Wykresy regresji lionwych', size=22)

names = weather_data.columns
colors=['indianred', 'firebrick', 'saddlebrown', 'olivedrab', 'darkolivegreen', 'seagreen', 'darkblue', 'navy', 'rebeccapurple']
c = 0
for i in range(3):
    for j in range(3):
        sns.regplot(y=daily_bike_traffic_data['Total'], x=weather_data[names[c]], ax=ax[i,j], ci=None, color=colors[c], label=f'y={regression_data_1["Total","a"][names[c]]:.3f}x+{regression_data_1["Total","b"][names[c]]:.3f}', scatter_kws={'alpha':0.6})
        ax[i,j].set_title('Reg. lin. dla łącznej il. rowerzystów w zależności od '+names[c][1]+' '+names[c][0])
        ax[i,j].set_xlabel(names[c][1]+' '+names[c][0])
        ax[i,j].set_ylabel('Łączna ilość rowerzystów')
        ax[i,j].legend()
        c+=1

plt.tight_layout(pad=5)
plt.savefig('correlations3_linreg_visualization.png')
plt.show()
```

```python
# Obliczenia i tablica
# YOUR CODE HERE
cols=(("Total","a"),("Total","b"),("Total","c"),("East","a"),
      ("East","b"),("East","c"),("West","a"),("West","b"),("West","c"))
columns = pd.MultiIndex.from_tuples(cols)
rows=(("Sky","Cloud cover"),("Precipitation","Total"))
indexes=pd.MultiIndex.from_tuples(rows)
regression_data_2=pd.DataFrame(None,index=indexes,columns=columns)

f=lambda x,a,b,c:a*np.exp(-b*x)+c


for i in ["Total","East","West"]:
        first=sp.optimize.curve_fit(f,weather_data["Sky","Cloud cover"],daily_bike_traffic_data[i])
        regression_data_2.loc[("Sky","Cloud cover"),i]=first[0]
                           
             
temp=weather_data["Precipitation","Total"].dropna()
l=[i for i in daily_bike_traffic_data.index.values if i in temp.index.values]
daily_bike_traffic_data_temp=daily_bike_traffic_data.loc[l]


idx=temp.sort_values(ascending=False).head(10).index
c_mean=daily_bike_traffic_data_temp.loc[idx].mean()
for i in ["Total","East","West"]:
        first=sp.optimize.curve_fit(f,temp,
                                    daily_bike_traffic_data_temp[i],
                                    bounds=((-np.inf,-np.inf,0.9*c_mean[i]),(np.inf,np.inf,1.1*c_mean[i])))
        regression_data_2.loc[("Precipitation","Total"),i]=first[0]
regression_data_2.style.format(precision=3)

#sum(weather_data["Precipitation","Total"].isna())
#raise NotImplementedError()
```
