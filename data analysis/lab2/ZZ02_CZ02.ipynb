Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

Indeks:

Kolumny:

Druga tablica (`recession_table_2`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego z rejonów geograficznych.

Indeks:
 - 1 poziomu: Region Name
 - 2 poziomu: Year (wszystkie lata dla każdego z rejonów)
 
Kolumny:
 - Kolejne przedziały (-25, -20], (-20, -15] ... (35, 40] zdefiniowane jako IntervalIndex

Trzecia tablica (`recession_table_3`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego stopnia zamożności.

Indeks:
 - 1 poziomu: Income Group
 - 2 poziomu: Year (wszystkie lata dla każdej z grup)
 
Kolumny:
 - Kolejne przedziały (-25, -20], (-20, -15] ... (35, 40] zdefiniowane jako IntervalIndex
 
Zapisz tablicę `recession_table_1` do pliku recession_table_1.pkl.

Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# Całość danych
cols = pd.IntervalIndex(pd.interval_range(start=-25, end=40, freq=5), name='IntervalIndex')
indx = pd.Index([2006,2007,2008,2009,2010,2011], name="Year")
recession_table_1=pd.DataFrame(index=indx,columns=cols)

data = np.zeros([6,13])
for i in range(2006,2012):
    dane=recession_data_1[str(i)].groupby(pd.cut(recession_data_1[str(i)], np.arange(-25,41,5))).count()
    data[i-2006]=dane/np.sum(dane)

recession_table_1 = pd.DataFrame(data, columns=cols,index=indx)
recession_table_1.to_pickle('recession_table_1.pkl')
recession_temp=recession_table_1.style.format(precision=2)
dfi.export(recession_temp, 'df2_new.png',table_conversion='matplotlib')
recession_temp
```

```python
# Podział na rejony
cols = pd.IntervalIndex(pd.interval_range(start=-25, end=40, freq=5), name='IntervalIndex')
indx = pd.MultiIndex.from_product([np.unique(recession_data_1['Region Name']),[2006,2007,2008,2009,2010,2011]], names=['Region Name','Year'])

data = np.zeros([30,13])
regions = np.unique(recession_data_1['Region Name'])
r=0
for i in range(5):
    for j in range(2006,2012):
        dane=recession_data_1[recession_data_1['Region Name']==regions[i]][str(j)].groupby(pd.cut(recession_data_1[str(j)], np.arange(-25,41,5))).count()
        data[r]=dane/np.sum(dane)
        r += 1

recession_table_2 = pd.DataFrame(data, columns=cols,index=indx)
recession_table_2.style.format(precision=2)

recession_temp=recession_table_2.style.format(precision=2)
dfi.export(recession_temp, 'df3.png',table_conversion='matplotlib')
recession_temp
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
cols = pd.IntervalIndex(pd.interval_range(start=-25, end=40, freq=5), name='IntervalIndex')
indx = pd.MultiIndex.from_product([np.unique(recession_data_1['Income Group']),[2006,2007,2008,2009,2010,2011]], names=['Income Group','Year'])

data = np.zeros([24,13])
groups = np.unique(recession_data_1['Income Group'])
r=0
for i in range(len(groups)):
    for j in range(2006,2012):
        dane=recession_data_1[recession_data_1['Income Group']==groups[i]][str(j)].groupby(pd.cut(recession_data_1[str(j)], np.arange(-25,41,5))).count()
        data[r]=dane/np.sum(dane)
        r += 1

recession_table_3 = pd.DataFrame(data, columns=cols,index=indx)
recession_table_3.style.format(precision=2)

recession_temp=recession_table_3.style.format(precision=2)
dfi.export(recession_temp, 'df4.png',table_conversion='matplotlib')
recession_temp

#raise NotImplementedError()
```

#### c) Dyskusja wyników

Zastanów się jakie są mocne i słabe strony poszczególnych typów wykresów wykonanych w tym zadaniu. W jakich przypadkach wykorzystanie konkretnych typów wykresów do wizualizacji może sprawiać problemy?

Diagramy częstości pokazują jak często występują poszególne wartości zaznaczone na osi x.
Nie pokazują one natomiast żadnej informacji na temat wartości naszych danych, więc należy je używać tylko wtedy, gdy interesuje nas tylko częstotliowść występowania jakichś zjawisk.

Jeśli chodzi o wykresy pudełkowe to dostarczają one nam ciekawych i użytecznych informacji statystycznych, które ciężko wyczytać z innych wykresów. Informują one głownie o tym, jak rozkładają się wartości dookoła mediany, więc jeśli jest to to, co nas interesuje, to jest to bardzo użyteczne narzędzie. Można z nich odczytać też kwantyle 1 i 3, wartości odstające i min/max wartości.

Violin ploty są można powiedzieć rozszerzeniem wykresów pudełkowych, ponieważ dostarczają nam podobnych informacji, a dodatkowo informują nas o gęstości prawdopodobieństwa naszych danych, więc jeżeli poza klasycznymi parametrami statystycznymi zawartymi w wykresach pudełkowych interesuje nas również gęstość prawdopodobieństwa, to należy skorzystać z violin plotu.

Strip ploty są dobry narzędziem jeśli chcemy sprawdzić jak wygląda rozkład częstości występowania poszególnych wartości naszej zmiennej. Punkty występują gęściej w miejscach, gdzie jest większa częstość występowania punktów, więc w niektórych przypadkach taki wynik będzie bardziej czytelny i łatwiejszy w odbiorze niż w przypadku np. diagramów częstości.

Swarm plot jest bardzo podobny do strip plotu z tą różnicą, że wartości nie nachodzą na siebie, więc czasami wykres może być bardziej czytelny, jednak to też zależy od zamierzonego efektu. W niektóych przypadkach większa gęstość i nachodzenie na siebie punktów może w bardziej czytelny sposób reprezentować większą czestotliwość występowania punktów.
