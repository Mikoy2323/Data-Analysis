#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE
#recession_table_1.loc["(-20, -15]"]
data=recession_table_1[["2006","2007","2008","2009","2010","2011",]]
descriptive_parameters_2=pd.DataFrame(np.zeros((6,6)),index=[["mean","std","Q1","Q2","Q3","mode"]],
                                      columns=data.columns)
descriptive_parameters_2.columns.name="Year"

#pomocniczne
half_indexes=(recession_table_1.index.left+recession_table_1.index.right)/2



#srednia
for j in recession_table_1.columns:
    temp=recession_table_1[j]
    mean=np.zeros(len(temp))
    for i in range(len(temp)):
        mean[i]=half_indexes[i]*temp.values[i]
    descriptive_parameters_2.loc["mean",j]=sum(mean)/sum(temp.values)

#std
for j in recession_table_1.columns:
    temp=recession_table_1[j]
    std=np.zeros(len(temp))
    m=descriptive_parameters_2.loc["mean",j]
    for i in range(len(temp)):
        std[i]=(half_indexes[i]-m)**2*temp.values[i]
    descriptive_parameters_2.loc["std",j]=np.sqrt(sum(std)/sum(temp.values))

    
#mode
for i in recession_table_1.columns:
    temp=recession_table_1[i]
    max_value=temp.max()
    max_index=list(temp).index(max_value)
    s=abs(temp.index[max_index].right-temp.index[max_index].left)
    mode=temp.index[max_index].left+s*(temp.values[max_index]-temp.values[max_index-1])/((temp.values[max_index]-temp.values[max_index-1])+(temp.values[max_index]-temp.values[max_index+1]))
    descriptive_parameters_2.loc["mode",i]=mode

#Q1
for j in recession_table_1.columns:
    temp=recession_table_1[j]
    N=temp.values.cumsum()[len(temp)-1]
    iN_4=1*N/4
    cs=temp.values.cumsum()
    for i in range(len(cs)-1):
        if cs[i+1]>=iN_4>=cs[i]:
            val=i+1
    L=temp.index[val].left
    C=abs(temp.index[val].left-temp.index[val].right)
    Q=L+(C/temp[val+1])*(iN_4-cs[val-1])
    descriptive_parameters_2.loc["Q1",j]=Q
    
#Q2
for j in recession_table_1.columns:
    temp=recession_table_1[j]
    N=temp.values.cumsum()[len(temp)-1]
    iN_4=2*N/4
    cs=temp.values.cumsum()
    for i in range(len(cs)-1):
        if cs[i+1]>=iN_4>=cs[i]:
            val=i+1
    L=temp.index[val].left
    C=abs(temp.index[val].left-temp.index[val].right)
    Q=L+(C/temp[val+1])*(iN_4-cs[val-1])
    descriptive_parameters_2.loc["Q2",j]=Q

    
#Q3
for j in recession_table_1.columns:
    temp=recession_table_1[j]
    N=temp.values.cumsum()[len(temp)-1]
    iN_4=3*N/4
    cs=temp.values.cumsum()
    for i in range(len(cs)-1):
        if cs[i+1]>=iN_4>=cs[i]:
            val=i+1
    L=temp.index[val].left
    C=abs(temp.index[val].left-temp.index[val].right)
    Q=L+(C/temp[val+1])*(iN_4-cs[val-1])
    descriptive_parameters_2.loc["Q3",j]=Q

descriptive_parameters_2


#raise NotImplementedError()
```
